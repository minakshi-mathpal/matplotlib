```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Minal\Desktop\cricket.csv")
```

```python
df[["score_india","score_pakistan"]].plot.hist(alpha=0.6)
```

```python
x= df["score_india"]
y = df["score_pakistan"]
```

```python
plt.hist([x,y])
plt.xlabel="runs"
plt.ylabel="Frequency"
plt.show()
```

```python
fig,axs=plt.subplots(1,2,sharex=True,sharey=True)
fig.suptitle("India Pak Comparison ")
axs[0].hist(x,color="orange")
axs[0].set_title="india"

axs[1].hist(y,color="green")
axs[1].set_title="pakistan"
```
